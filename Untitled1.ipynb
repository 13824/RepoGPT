```python
import torch
from torch.utils import data
import pandas as pd
import numpy as np
import sklearn
import scipy
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('/content/sample_data/california_housing_train.csv')
test_df = pd.read_csv('/content/sample_data/california_housing_test.csv')
longitude = torch.tensor(train_df['longitude'].values, dtype = torch.float32)
latitude = torch.tensor(train_df['latitude'].values, dtype = torch.float32)
age = torch.tensor(train_df['housing_median_age'].values, dtype = torch.float32)
rooms = torch.tensor(train_df['total_rooms'].values, dtype = torch.float32)
bedrooms = torch.tensor(train_df['total_bedrooms'].values, dtype = torch.float32)
population = torch.tensor(train_df['population'].values, dtype = torch.float32)
households = torch.tensor(train_df['households'].values, dtype = torch.float32)
income = torch.tensor(train_df['median_income'].values, dtype = torch.float32)
price = torch.tensor(train_df['median_house_value'].values, dtype = torch.float32)
testprice = torch.tensor(test_df['median_house_value'].values, dtype = torch.float32)
```

```python
def load_array(data_arrays, batch_size):
  return data.DataLoader(data_arrays, batch_size, shuffle=True)
data_iter = load_array((longitude, latitude, age, rooms, bedrooms, population, households, income, price), 1000)
```

```python
net = torch.nn.Linear(8, 1)
net.weight.data.fill_(0)
net.bias.data.fill_(torch.mean(price))

loss = torch.nn.MSELoss()
trainer = torch.optim.SGD(net.parameters(), lr=0.0000000000009)
```

```python
num_epochs = 2000
epochlist = []
losslist = []


for epoch in range(num_epochs):

  for lon, lat, a, r, bed, pop, h, i, p in data_iter:

    features = torch.stack((lon, lat, a, r, bed, pop, h, i), dim = 1)

    l = loss(net(features), p.unsqueeze(1))

    trainer.zero_grad()

    l.backward()

    trainer.step()
    epochlist.append(epoch+1)
    losslist.append(l.item())

    print(f'epoch {epoch + 1}, loss {l.item():f}')
plt.plot(epochlist, losslist)
plt.show()
```

```python
x = []
trainlist = []
residual = 0
for t in range(0, test_df.shape[0]):
    test_row = test_df.iloc[t]
    test_features = test_row[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]
    test_features_tensor = torch.tensor(test_features.values, dtype=torch.float32)
    test_features_tensor = test_features_tensor.unsqueeze(0)
    residual = (test_row["median_house_value"] - net(test_features_tensor).item())
    trainlist.append(net(test_features_tensor).item())
    x.append(t)
print(1-torch.mean(testprice)/residual)
plt.subplot(1, 2, 1)
plt.plot(x, trainlist)
plt.subplot(1, 2, 2)
plt.plot(x, test_df["median_house_value"])
plt.show()
```
